```python
counts = movies['movie'].value_counts().reset_index()
print(counts)
columns = ["movie", "counts"]
plt.xlabel("Movies")
plt.ylabel("Counts")
plt.xticks(rotation=45, ha='right')
sns.barplot(data=counts,x="movie", y='count', hue='movie',palette="Set2")
```

```python
counts = movies['rating'].value_counts().reset_index()
print(counts)
plt.xlabel("Ratings")
plt.ylabel("Counts")
sns.barplot(data=counts,x='rating', hue='rating', y='count', palette="Set2")
```

```python
user_movies = movies.pivot_table(index='userId',columns='movie',values='rating')
user_movies
```

```python
user_movies.fillna(0, inplace=True)
```

```python
user_movies
```

```python
from sklearn.metrics import pairwise_distances
```

```python
user_sim = 1-pairwise_distances(user_movies.values,metric='cosine')
user_sim
```

```python
np.fill_diagonal(user_sim, 0)
user_sim
```

```python
user_sim.shape
```

```python
user_sim = pd.DataFrame(user_sim)
user_sim
```
